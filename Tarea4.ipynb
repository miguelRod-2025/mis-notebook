## Temas Pila (LIFO) Historial con capacidad limitada. Los navegadores modernos limitan el historial para no consumir memoria indefinidamente.

## Implementa una clase NavegadorWeb con las siguientes características:

### 1. El constructor recibe un parámetro CAPACIDAD (número máximo de páginas que puede guardar el historial). Si el historial está lleno al visitar una página nueva, se elimina la página más antigua para hacer espacio.

### 2. visitar(url): agregar la página actual al historial y navega a la nueva URL. Si el historial ya alcanzó su capacidad, descarta la página más antigua antes de agregar la nueva.

### 3. atras(): regresa a la página anterior si existe historial: de lo contrario, muestra un mensaje indicando que no hay paginas anteriores.

### 4. pagina_actual (): imprime la URL en la que se encuentra el navegador.

### 5. mostrar_historial(): imprime todas las páginas guardadas, de la más reciente a la más antigua, sin modificar la pila.

## Explicación escrita esperada: justifica por qué una pila para el historial y explica cómo implementas el límite de capacidad sin perder el comportamiento LIFO.

```python
# El constructor llamado Navegador Web

class NavegadorWeb:
  """
  Implementa un navegador web con historial limitado.
  """
  def __init__(self, capacidad):
    self.capacidad = capacidad   # aquí mandamos el número de página número de páginas
    self.historial = []   # vamos a guardar aquí la capacidad, que funciona como pila de una lista
    self.actual = None


# Vamos agregando las visitas a las paginás web definimos visitar(url)
  def visitar(self, url):
    if self.actual:
        if len(self.historial) >= self.capacidad:
            # Aquí vamos a eliminar la más antigua, que fue la primera en entrar que fue en el indice 0
            self.historial.pop(0)
        self.historial.append(self.actual)  # recorremos la pila y actualizamos
    # vamos a ala página nueva
    self.actual = url
    print(f"Visitamos la url: {url}")


# Quitamos el último elemento (tope) y lo devolvemos
# vamos a regresar a la página anterior si ya tenemos almacenado las visitas
  def atras(self):
      if self.isEmpty():
          print("No hay páginas anteriores")  # verificamos que no hay páginas
          return None

      # regresamos la última página
      self.actual = self.historial.pop()
      print(f"Regresamos a la página: {self.actual}")  # seguimos verificando que se cumpla
      return self.actual


# imprimimos la página visitada y la guardamos se usa como el peek
  def pagina_actual(self):
      if self.actual:
        print(f"La página actual es: {self.actual}")
      else:
        print("No hay páginas visitadas")


# Mostrar el historial
  def mostrar_historial(self):
      print("Historial (reciente al antiguo):")
      for pagina in reversed(self.historial):
          print(pagina)


# Verificamos si no hay elementos
  def isEmpty(self):
    return len(self.historial) == 0


pila = NavegadorWeb(capacidad=5)

# vamos apilando pero aqui vamos a pilar la visita a páginas web
pila.visitar("www.google.com")
pila.visitar("www.facebook.com")
pila.visitar("www.twitter.com")
pila.visitar("www.instagram.com")
pila.visitar("www.youtube.com")
print()
pila.pagina_actual()
print()
pila.mostrar_historial()
print()
pila.atras()
print()
pila.pagina_actual()
print()
pila.mostrar_historial()
```

## En clase estudiamos cómo una pila permite evaluar expresiones en notación postfija (por ejemplo, 3 5 + equivale a 3 + 5, y 3 5 + 2 * equivale a (3 + 5) × 2). En este ejercicio implementarás un evaluador completo que soporte los cuatro operadores aritméticos básicos. Escribe una función evaluar_postfija(expresion) que reciba una cadena de texto con tokens separados por espacios y devuelva el resultado numérico. La función debe:

### 1. Recorrer cada token de la expresión.

### 2. Si el token es un número (entero o decimal), apilarlo.

### 3. Si el token es un operador (+,-, *, /), desapilar los dos últimos operandos, aplicar la operación y apilar el resultado. El primer valor desapilado corresponde al operando derecho.

### 4. Al finalizar, el único elemento que queda en la pila es el resultado.

### 5. Si la expresión está mal formada (pila vacía al intentar desapilar, o más de un elemento al terminar), lanzar un ValueError con un mensaje descriptivo.

## Explicación escrita esperada: descibir el papel de la pila en cada paso del algoritmo. ¿Porque el primer valor desapilado es el operador derecho y no el izquierdo?

```python
# Ejercicio 2

def evaluar_postfija(expresion):
    """
    Evaluar una expresión en notación postfija.
    """
    pila = []  # esta pila la usamos para los operandos
    operadores = {'+', '-', '*', '/'}


    for token in expresion.split():
        if token not in operadores:   # verificamos que si el token no
            # si el token es un número apilamos
            try:
                numero = float(token)  # usamos un cast en caso de un decimal
                pila.append(numero)  # lo vamos agregando
                print(f"Apilamos: {numero}")
            except ValueError:
                raise ValueError(f"Token inválido: {token}")
        else:
          # ahora vamos con los operadores con la condición
          if len(pila) < 2:
              raise ValueError("Expresión inválida: no hay suficientes operandos.") #
          derecha = pila.pop()  # comenzamos a revisar por la derecha
          izquierda = pila.pop()  # por la izquierda

          if token == '+':
              resultado = izquierda + derecha
          elif token == '-':
              resultado = izquierda
          elif token == '*':
              resultado = izquierda * derecha
          elif token == '/':
              if derecha == 0:
                  raise ValueError("División por cero.")
              resultado = izquierda / derecha

          pila.append(resultado)
          print(f"Apilamos: {resultado}")

    if len(pila) != 1:
        raise ValueError("Expresión inválida: no se puede determinar el resultado.")

    return pila[0]



# implementar los cuatro operadores aritméticos

expresion = "3 5 2 * +"  # equivale a 3 + (5 * 2)
resultado = evaluar_postfija(expresion)
print(f"El resultado de la expresión {expresion} es: {resultado}")
```

## En clase,

```python
# Ejercicio 3  colas

class SistemaPrioridad:
    def __init__(self):
        self.cola_urgentes = []
        self.cola_normales = []


    # creamos el método o función para encolar enqueue
    def agregar(self, nombre, descripcion, urgente=False):
        """
        Agregamos a la cola de pacientes
        """

        if urgente:
            self.cola_urgentes.append((nombre, descripcion))
            print(f"Se agregó a la cola de urgentes: {nombre} - {descripcion}")
        else:
            self.cola_normales.append((nombre, descripcion))
            print(f"Se agregó a la cola de normales: {nombre} - {descripcion}")

    # creamos el metodo para desencolar dequeue
    def atender_siguiente(self):
        """
        atendemos primero la solicitud de urgencia
        """

        if self.cola_urgentes:
            nombre, descripcion = self.cola_urgentes.pop(0)
            print(f"Atendida la solicitud de urgencia: {nombre} - {descripcion}")
            return (nombre, descripcion)

        elif self.cola_normales:
            nombre, descripcion = self.cola_normales.pop(0)
            print(f"Atendida la solicitud normal: {nombre} - {descripcion}")
            return (nombre, descripcion)
        else:
            raise ValueError("No hay pacientes en las colas")
            return None


  # imprimir cuantas solicitudes hay en cada cola
    def mostrar_estado(self):
      """
      imprimir cuantas solicitudes hay en cada cola
      """
      print(f"Solicitudes en la cola de urgentes: {len(self.cola_urgentes)}")
      print(f"Solicitudes en la cola de normales: {len(self.cola_normales)}")

sistema = SistemaPrioridad()
print()

sistema.agregar("paciente 1", "consulta médica", urgente=False)
sistema.agregar("paciente 2", "cirugía", urgente=False)
sistema.agregar("paciente 3", "examenes", urgente=False)
sistema.agregar("paciente 4", "dolor", urgente=True)
print()

sistema.mostrar_estado()
print()

sistema.atender_siguiente()
sistema.atender_siguiente()
print()

sistema.mostrar_estado()

```

## En clase simulamos una cola de impresión básica. En este ejercicio la extenderás para que lleve un registro de los documentos ya impresos y pueda generar un reporte de uso. Implementa una clase ImpresoraAvanzada con las siguientes características:
## 1. El constructor recibe la velocidad de impresión en páginas por minuto.
## 2. agregar(nombre, paginas): encola el documento e imprime un mensaje de confirmación.

## 3. imprimir_siguiente(): desencola e imprime el primer documento. Guarda en una lista interna _historial el nombre, el número de páginas y el tiempo de impresión (paginas / velocidad).
## 4. imprimir_todos(): llama a imprimir_siguiente() hasta que la cola esté vacía.
## 5. reporte(): imprime: Total de documentos impresos. Total de páginas impresas. Tiempo total de impresión (en minutos). El documento con más páginas y el documento con menos páginas.

```python
# Ejercicio 4

# clase
class ImpresoraAvanzada:
    def __init__(self, velocidad_paginas_por_minuto):
          self.cola_impresos = []   # guardamos los documentos pendientes
          self.velocidad = velocidad_paginas_por_minuto   #
          self._historial = []  # guardamos los documentos impresos

    # recibe la velocidad de impresión página por minuto.
    def agregar(self, nombre, paginas):
          self.cola_impresos.append((nombre, paginas))
          print(f"Se encoló el documento: {nombre}  ({paginas} páginas)")

    # imprimir el primer documento
    def imprimir_siguiente(self):
          if self.cola_impresos:
              nombre, paginas = self.cola_impresos.pop(0)  # vamos imprimiendo y borrando
              tiempo = paginas / self.velocidad * 60  # tiempo en segundos
              self._historial.append((nombre, paginas, tiempo))
              print(f"Imprimiendo el documento: {nombre} ({paginas} páginas, {tiempo:.2f} segundos)")
          else:
              print("No hay documentos en la cola para imprimir.")

    # llama al siguiente en imprimir y va vaciando la cola
    def imprimir_todos(self):
          while self.cola_impresos:
              self.imprimir_siguiente()
          print("No hay más documentos en la cola.")

    # imprimimos los detalles de los documentos
    def imprimir_detalles(self):
          print("Detalles de los documentos impresos:")
          for nombre, paginas, tiempo in self._historial:
              print(f"Nombre: {nombre}, Páginas: {paginas}, Tiempo: {tiempo:.2f} segundos")

    # para imprimir el reporte
    def reporte(self):
        if not self._historial:
          print("No hay documentos impresos.")
          return

        total_docs = len(self._historial)
        total_paginas = sum(paginas for _, paginas, _ in self._historial)
        total_tiempo = sum(tiempo for _, _, tiempo in self._historial)
        doc_mas_paginas = max(self._historial, key=lambda x: x[1])
        doc_menos_paginas = min(self._historial, key=lambda x: x[1])

        print(" === Reporte de Impresión === ")
        print(f"Total de documentos impresos: {total_docs}")
        print(f"Total de páginas impresas: {total_paginas}")
        print(f"Tiempo total de impresión: {total_tiempo:.2f} minutos")
        print(f"Documento con más páginas: {doc_mas_paginas[0]} ({doc_mas_paginas[1]} páginas)")
        print(f"Documento con menos páginas: {doc_menos_paginas[0]} ({doc_menos_paginas[1]}páginas)")


impresora = ImpresoraAvanzada(velocidad_paginas_por_minuto=10)

impresora.agregar("Documento 1", 15)
impresora.agregar("Documento 2", 20)
impresora.agregar("Documento 3", 10)
print()
impresora.imprimir_todos()
print()
impresora.imprimir_detalles()
print()
impresora.reporte()

```

## En este ejercicio combinarás lo aprendido sobre pilas y colas para resolver un problema que requiere razonar sobre cuál estructura conviene en cada operación. Implementa un TDA PalabraMagica que reciba una cadena de texto en el constructor y ofrezca los siguientes métodos. Restricción: antes de cargar los caracteres en la estructura, elimina los espacios en blanco y convierte todo a minúsculas. Luego, cárgalos en una pila y/o cola usando un ciclo for. A partir de ese punto, usa únicamente las operaciones del TDA (push, pop, enqueue, dequeue, etc.) para resolver cada método. No se permite usar reversed(), slicing inverso ([::-1]), ni acceso por índice (cadena[i]) fuera del ciclo de carga inicial.

# 1. es_palindromo() → devuelve True si la cadena es un palíndrome (se lee igual de izquierda a derecha que de derecha a izquierda), ignorando mayúsculas y espacios.
# 2. invertir() → devuelve la cadena original al revés como una nueva cadena.
# 3. es_anagrama(otra_cadena) → devuelve True si la cadena original es un anagrama de otra_cadena (mismas letras, diferente orden), ignorando mayúsculas y espacios.

```python
# Ejercicio 5

# implementar un TDA

class palabraMagica:
    def __init__(self, cadena):
      self.cadena = cadena.lower().replace(" ", "") # "" quitamos los espacios y lower vuelve minusculas a las mayusculas

    # un aplindromo se lee de izq. a der. y der a izq. lo mismo
    def es_palindromo(self):
        """
        Debe de devolver True si la cadena es un palíndrome (ya que se lee igual
        de izquierda a derecha que de derecha a izquierda), ignorando mayúsculas y espacios,
        comparamos manualmente el índice.
        Desde mi forma de ver es más fácil usar la pila por el simple hecho
        de que como va entrando digamos al índice 0, índice 1 y así sucesivamente
        y recordando el método divide y venceras, solo se va comparando los indices
        si coinciden entonces es un palíndromo.
        """
        letras = []
        for letra in self.cadena:
            letras.append(letra)  # agregamos elemento

          # vamos a comparar elementos por ejemplo del índice 0 con el penúltimo el 1 con el antepenúltimo y así sucesivamente
        inicio = 0     # inicializamos por supuesto en cero
        fin = len(letras) - 1  # aquí con el penúltimo
        while inicio < fin:
              if letras[inicio] != letras[fin]:  # este funciona como un auxuliar para comparar
                  return False
              inicio += 1  # recorremos
              fin -= 1     # recorremos
        return True


    # invertir cadena
    def invertir(self):
        """
        Debe devolver la cadena original al revés como una nueva cadena.
        Usando lo mismo que mencione anteriormente usamos la pila para invertir la cadena
        quitando moviendo y poniendo los elementos.
        Usamos el for para recorrer en esté caso la cadena y con el
        while para invertirla.
        """

        pila = []
        for letra in self.cadena:
            pila.append(letra)

        invertida = ""
        while pila:
            invertida += pila.pop()

        return invertida  # return fuera del bucle

    # anagrama mezclar las palabras para formar otra
    def es_anagrama(self, otra_cadena):
      """
      Debe devolver, True si la cadena es un anagrama de otra cadena
      Use sorted para ordenar las secuencia de la cadena, desde mi idea
      es tomar algo ordenado revolverlo y volverlo a ordenar y comparar.
      """
      otra = otra_cadena.lower().replace(" ", "")

      # sorted se toma una secuencia como un diccionario y devuelve una nueva secuencia ordenada
      return sorted(self.cadena) == sorted(otra)


p1 = palabraMagica("Anita lava la tina")
print("Es palindrome ", p1.es_palindromo())   # True
print("Si se invierte ", p1.invertir())       # "anitalavalatina"
print()
p3 = palabraMagica("Dábale arroz a la zorra el abad")
print("es palindromo: ", p3.es_palindromo())        # False
print()
p4 = palabraMagica("La ruta natural")
print("es palindromo: ", p4.es_palindromo())
print()


p2 = palabraMagica("Amor Roma")
print("es palindromo: ", p2.es_palindromo())        # False
print("Si se invierte: ", p2.invertir())            # "roma"
print("Si es anagrama: ", p2.es_anagrama("Roma Amor"))    # True
print("Si es anagrama: ", p2.es_anagrama("mora"))    # True
print("Si es anagrama: ", p2.es_anagrama("casa"))    # False
print()
```
