# Ejercicio 1 clasificador de números
## Escriba un programa que solicite al usuario ingresar un número entero y muestre la siguiente información:

## 1. Si el número es positivo, negativo o cero; 2. Si el número es par o impar. 3. si el número es múltiplo de 3.

```python
# Ejercicio 1 Clasificador de número
# checar si un número es positivo, negativo cero; si es par e impar y por último múltiplo de 3
numero = int(input("Ingrese un número entero: "))

if numero > 0 :
      print("El número es positivo")
elif numero < 0:
      print("El número es negativo")
else:
      print("El número es cero")

if numero % 2 == 0:
      print("El número es par")
else:
      print("El número es impar")

if numero % 3 == 0:
      print("El número es múltiplo de 3")
```

# Ejercicio 2 Adivina el número
## Escriba un programa que elija un número secreto fijo y le de al usuario hasta 5 intentos para adivinarlo. En cada intento el programa debe indicar si el número secreto es mayor o  menor al ingresado.

## Si el usuario adivina, se muestra un mensaje de felicitación indicando en cuántos intentos lo logró. Si agota los intentos sin acertar, se revela el número secreto.

```python
# Ejercicio 2 : Adivina el número

numero = 72  # número a comparar
adivino = False  # yo lo conozco como bandera
intentos = 0  # Aquí inicializamos el contador


# Hacemos el ciclo; es verdadero = intentos y adivino falso = Falso
while intentos < 5 and adivino == False:
  intentos = intentos + 1  # vamos aumentando a el contador por cada intento
  numero_ingresado = int(input("Ingrese un número: "))

  if numero_ingresado == numero:  # Aquí hacemos la comparación
    adivino = True  # Si acierta cambia la bandera a verdadero y ya quedo :)
    print("Adivinaste el número en ", intentos, " intentos")
    print("¡Felicidades!")

  elif numero_ingresado > numero:
    print("El número secreto es menor, intento: ", intentos)
  else:
 # elif numero_ingresado < numero:
    print("El número secreto es mayor, intento ", intentos)

if adivino == False: # Aquí lo negamos para salir
    print("Has agotado tus intentos, el número secreto era: ", numero)

```

# Ejercicio 3: Validador de contraseña
## Escribe un programa que solicita al usuario ingresar una contraseña y verifique si cumple con las siguientes reglas.
## Tiene al menos ocho caracteres, contiene al menos una una letra mayúscula y contiene al menos un dígito.
## El programa debe indicar para cada regla si cumple o no, y al final mostrara si la contraseña es válida, solo si cumple las tres reglas.

```python
# Ejercicio 3: Validador de contraseña

# La contraseña debe contener por lo menos 8 caracteres una mayuscula un dígito según estas reglas

contraseña = input("Ingrese una contraseña, que por lo menos tenga una Mayuscula,\n un dígito y 8 caracteres:\n ")
print()
if len(contraseña) < 8: # sabemos que len es la longitud
  print("La contraseña debe tener al menos 8 caracteres, conytraseña incorrecta")

elif not any(char.isdigit() for char in contraseña):
  print("Lacontraseña debe contener al menos un dígito, contraseña incorrecta")

elif not any(char.isupper() for char in contraseña):
  print("La contraseña debe contener al menos una letra mayúscula, contraseña incorrecta")
  print()
else:
  print("Contraseña valida!!")

```

# Ejercicio 4: Años bisiestos en un intervalo
## Escriba un programa que:
## 1. Solicitar al usuario ingresar dos años enteros: a) año inicial b) año final; 2. Muestre en pantalla todos los años bisiestos comprendidos entre esos dos años, incluyendo ambos extremos. 3. Al finalizar, indique cuántos años bisiestos se encontraron en total.
## Definición de año bisiestos: a) es divisible entre 4 y no es divisible entre 100, b) o es divisible entre 400
# Aclaraciones; si el año inicial es mayor que el año final, el programa debe manejar la situación adecuadamente (por ejemplo, mostrando un mensaje de error o intercambiando los valores). Ambos años introducidos deben considerarse dentro del intervalo. Solo se deben avaluar años enteros.

```python
from typing_extensions import final
# Ejercicio 4: Año bisiesto en un intervalo

# Pedimos al usuario ingrese los datos
año_inicial = int(input("Ingrese el año inicial: "))
año_final = int(input("Ingrese el año final: "))

# intercambio de los años en el caso de que el año inicio sea mayor
if año_inicial > año_final:
  año_inicial, año_final = año_final, año_inicial
  print("Los años han sido intercambiados", ) # esta parte la use para ir probando el intercambio

# Contador de años bisiestos
contador_bisiestos = 0

# Recorremos con for [el intervalo]
for año in range(año_inicial, año_final+1):
  if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    print(año, "es un año bisiesto")
    contador_bisiestos += 1

# Imprimimos el contador
if contador_bisiestos == 0:
  print("No hay años bisiestos en ese intervalo")
else:
  print("Hay", contador_bisiestos, "años bisiestos en ese intervalo")

```
