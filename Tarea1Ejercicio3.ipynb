```python
# Gasto por viaje
litrosPorKilometro = 0  # inicializamos en cero
combustibleTotal = 0    # inicializamos en cero
costoPorKilometro = 0   # inicializamos costo por kilometro

# Solicitamos al usuario los kilometros recorridos
print("Me puede proporcionnar los kilometros recorridos")
kilometros = float(input())

# Solicitamos los litros de combustible consumidos
print("Me puede proporcionar los litros que se gasto")
litros = float(input())

# Solicitamos el precio por litro de combustible
print("Me puede proporcionar el precio de la gasolina por litro")
precio_litro = float(input())

# Calculamos y devolvemos: (consumo en km por lt, costo total de combustible, costo por km)

# Calculamos cuantos litros por kilometro se gasto
litrosPorKilometro = (kilometros/litros)
print(f"Usted recorrio {litrosPorKilometro: .4f} kilometro por cada litro")

# Calculamos costo Total del combustible
combustibleTotal = (litros * precio_litro)
print(f"Usted gasto  {combustibleTotal: .2f} en gasolina")

# Calculamos el precio por kilometro
costoPorKilometro = (kilometros/precio_litro)
print(f"A usted le salio {costoPorKilometro: .4f} por kilometro recorrido")

```
